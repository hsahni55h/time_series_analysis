```python
# Import necessary modules for statistical analysis and plotting
import scipy.stats
import pylab
```

scipy.stats.probplot(df_comp.spx, plot=pylab): Generates a Q-Q (Quantile-Quantile) plot, which is used to compare the distribution of a dataset to a theoretical distribution (in this case, a normal distribution). The plot helps in assessing whether the data follows a normal distribution.

QQ Plot: A QQ plot is a graphical tool to help assess if a dataset follows a given distribution (usually normal). If the data points fall approximately along a straight line in the plot, it suggests that the data follows the specified distribution.

```python
# Create a QQ plot to check if the data follows a normal distribution
scipy.stats.probplot(df_comp.spx, plot=pylab)
plt.title("QQ Plot", size=24)
pylab.show()
```
